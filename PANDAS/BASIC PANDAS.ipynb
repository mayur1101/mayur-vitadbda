# HANDLE MISSIND DATA


```python
import pandas as pd
df=pd.read_csv("d://weather_data.csv")
df
```

```python
df.fillna(0)            #it will replace all NAN values with 0in orignal dataframe
```

```python
df
```

```python
new_df=df.fillna(0)          #changes in new_df
new_df
```

```python
new_df=df.fillna({'temperature':0,
                  'windspeed':0,                  #fillinf missing values with respective columns.
                  'event':"no event"})
new_df
```

```python
new_df=df.fillna(method="ffill")    #ffill->forward fill,If there is null values and to set values of previous row.
new_df                                            
```

```python
new_df=df.fillna(method="bfill")    #ffill->backward fill,If there is null values and to set values of next row.
new_df                                            
```

```python
new_df=df.fillna(method='ffill',limit=1)  #It will replace NAN values with only limit 1 other values will pe NAN.
new_df
```

```python
new_df=df.interpolate()    #It will print the integer guessing right values i.e. integer & not string
new_df
```

```python
new_df=df.dropna()       #It will drop all records of NA
new_df
```

```python
new_df=df.dropna(how='all')  #Drop all rows with all null values.
new_df
```

```python
new_df=df.dropna(thresh=1)   #It will drop rows which doesnot have atleast one entry in row.
new_df
```

# GROUP BY FUNCTION

```python
#One groupby object is created we can calculate various statistics for each group.Creating groupby object only verifies
#that you have passed a valid mapping.By default group keys are sorted during the group by operation.
#Group by first divide into groups then apply analytics to each group and then aggregate & then combine and make dataframes.
#It is known as "split","apply","combine".
import pandas as pd
df=pd.read_csv("d://city.csv")
df
```

```python
g=df.groupby('CITY')
g
```

```python
for CITY,CITY_DF in g:
    print(CITY)
    print(CITY_DF)
```

```python
#What is the max salary in each of city.
g.max()
```

```python
g.mean()
```

```python
g.get_group('MUMBAI')
```

```python
g.describe()
```

```python
%matplotlib inline
g.plot()
```

```python
india_weather=pd.DataFrame({
    "city":["mumbai","delhi","bangalore"],
    "temperature":[32,45,30],
    "humidity":[80,60,78]
})

us_weather=pd.DataFrame({
    "city":["newyork","chicago","orlando"],
    "temperature":[32,45,30],
    "humidity":[80,60,78]
})

df=pd.concat([india_weather,us_weather])
df
```

```python
df=pd.concat([india_weather,us_weather],axis=1)
df
```

```python
df=pd.concat([india_weather,us_weather],ignore_index=True)
df
```

```python
df=pd.concat([india_weather,us_weather],keys=["india","us"])
df
```

# PANDAS MERGE


```python
df1=pd.DataFrame({
    "city":["chicago","newyork","orlando"],
    "temperature":[32,45,30],
    "humidity":[80,60,78]
})

df2=pd.DataFrame({
    "city":["newyork","chicago","orlando"],
    "temperature":[32,45,30],
    "humidity":[80,60,78]
})
df3=pd.merge(df1,df2,on="city")
df3
```

```python
df=pd.merge(india_weather,us_weather,on="city",how="outer")   #union of two frames
df
```

```python
df=pd.merge(india_weather,us_weather,on="city",how="left")   #left records
df
```

```python
df=pd.merge(india_weather,us_weather,on="city",how="right")   #right records
df
```

```python
df=pd.merge(india_weather,us_weather,on="city",how="outer",indicator=True)   #union of two frames
df
```

```python
df=pd.merge(india_weather,us_weather,on="city",suffixes=('_left','_right'))   #common columns of both dataframes
df
```
